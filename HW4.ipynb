## Part 1: Loading data and cleaning (20 points)
1) Store the name of your file in `filepath` and open it with the `fileobj` variable. Use the `readline()` function and a for loop, find out how many lines of header are in the "Pakistan_sea_level_psmsl.txt" file and store this number in the `header_lines` variable. Read the information in the header. In this data file, missing sea level data points are filled with a placeholder "masking" value. (5 points)
>1) What is the mask value for missing data in this dataset? Answer this question in the markdown cell below the solution cell.

2) Use numpy `genfromtext()` to read each column of your text file into a Numpy array as floats and store the array in the `sea_level` variable.(_HINT: Remember that indexing starts with 0 in Python!_). Store the columns as individual 1-D arrays (excluding "number of missing days") in the provided answer variables, `year_month`, `depth_mm`, `data_flag`. (5 points)
>1) What is the delimiter type for this text file? Answer this question in the markdown cell below the solution cell.

3) Using the numpy `where()` function and its first positional argument ([Documentation](https://numpy.org/doc/stable/reference/generated/numpy.where.html)), find the indices of the masked depth values that represent missing values in your Numpy array and store the resulting array in `mask_inds`. Using `mask_inds`, replace the masking values in `depth_mm` with `np.nan` values. (5 points)

4) PSMSL also flags data that is likely erroneous. In this data, a data quality flag value of 11 represents bad data. We will also replace these suspicious values with `np.nan` values to ignore them. (5 points)

  >a) Find the indices of bad data quality values in `data_flag` by numpy `where()` and store in `bad_inds`.

  >b) Save the corresponding "year" and "depth_mm" values of bad data in a separate arrays, `bad_years` and `bad_depths` respectively. 
  
  >c) Replace bad depth values in `depth_mm` with `np.nan` values.

````python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy import interpolate 
from scipy import stats

# Defined file path and count header lines
filepath = 'Pakistan_sea_level_psmsl.txt'

header_lines = 22  
mask_value = -99999.0 # Here is the mask value

# Loaded data with numpy.genfromtxt
sea_level = np.genfromtxt(filepath, skip_header=header_lines, delimiter=';', dtype=float) # The delimiter type is ;

# Extracted columns into variables
year_month = sea_level[:, 0]  
depth_mm = sea_level[:, 1]   
data_flag = sea_level[:, 3]   

# Replaced masking values with np.nan
mask_inds = np.where(depth_mm == mask_value)
depth_mm[mask_inds] = np.nan

# Replaced bad data flagged as 11 with np.nan
bad_inds = np.where(data_flag == 11)
badyears = year_month[bad_inds] 
baddepths = depth_mm[bad_inds]  
depth_mm[bad_inds] = np.nan

# Printed the data out
print(f"Number of header lines: {header_lines}")
print(f"Masking value for missing data: {mask_value}")
print("First 12 bad data years:", list(badyears[:12]))  
print("First 12 bad data depths:", list(baddepths[:12]))  
````

````python

print('Part 1: ')

# Transposed the list of arrays
list_data = np.array([year_month, depth_mm, data_flag]).T

# Created the DataFrame
df = pd.DataFrame(data=list_data, columns=["Year", "Depth", "Flag"])

# Displayed the DataFrame
print("Initial DataFrame:")
print(df)

print('Part 2: ')
# Found the number of missing rows
missing_rows = len(df.loc[df["Depth"].isnull()])

# Did the missing interpolated values
df["Depth_interp"] = df["Depth"].interpolate()

# Displayed the number of missing rows and the updated DataFrame
print("There are", missing_rows, "missing data points.")
print("\nDataFrame after interpolation:")
print(df)

print('Part 3: ')
# Plotted 
fig, axes = plt.subplots(nrows=2, figsize=(10, 6), sharex=True)

# Top subplot for the original depth with bad data
axes[0].plot(df["Year"], df["Depth"], label="Original Data", marker='o', linestyle='-', color='blue')
axes[0].plot(badyears, baddepths, label="Bad Data", marker='x', linestyle='none', color='red')
axes[0].set_title("Original Depth Data with Bad Values")
axes[0].set_ylabel("Depth (mm)")
axes[0].legend()

# Bottom subplot for the Interpolated depth
axes[1].plot(df["Year"], df["Depth_interp"], label="Interpolated Data", marker='o', linestyle='-', color='green')
axes[1].set_title("Interpolated Depth Data")
axes[1].set_xlabel("Year and Month")
axes[1].set_ylabel("Depth (mm)")
axes[1].legend()

# layout and display
plt.tight_layout()
plt.show()

# d) The interpolated provides a reasonable estimate of the missing depth data trends but has limitations like not fullly capturing the natural variability of sea level change due to the assumption of a straight-line trend.
````

>a) To better visualize yearly trends, calculate the rolling yearly mean of depths for "Depth" column using Pandas `rolling()` function ([Documentation](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.rolling.html)). _HINT: The data is in monthly resolution. How many months will you need to get the rolling yearly mean?_

>b) Store the rolling mean depth values in a new column called "yearly_height" in `df`.

>d) Plot your "yearly_height" values against "Year" on the **top** subplot, overlaying the original time series data. Make this line a different color and line style, and add it to the legend in the top subplot.


2) The `rolling()` function does not deal with nan values, which represents our missing and suspicious data. This results in creating larger data gaps in the calculated rolling mean depth values compared to the original data. Here, we will _interpolate and extrapolate_ the rolling mean values to resolve the data gaps. (10 points)

>a) Apply the `interpolate()` function to the `df` to the df["yearly_height"] column. Set the argument `limit_direction = "both"` to extrapolate the data at the ends. Store your interpolated depth values in a new column called "Depth_interp_roll" in `df`.

>b) Building off your plot from Part 1 above, plot your interpolated rolling mean depth values ('Depth_interp_roll') on your **bottom** subplot as a different color. Add it to the legend in the bottom subplot. (You do not need to copy/paste the code here, just add the line of code underneath your plotting code from Part 1).

>c) Do you think this interpolation describes the trends in missing depth data more or less accurately than your first interpolation attempt? Why or why not? Store your answer in the markdown cell below the solution cell. 

````python

print("Part 4:")
# Ensured rolling handles NaN values appropriately
df["yearly_height"] = df["Depth"].rolling(window=12, min_periods=12).mean()

# Checked for NaN values in yearly_height
print("NaN count in yearly_height:", len(df.loc[df["yearly_height"].isnull()]))

# Interpolate and extrapolate the rolling mean depth values
df["Depth_interp_roll"] = df["yearly_height"].interpolate(limit_direction="both")

# Checked for NaN values in Depth_interp_roll
print("NaN count in Depth_interp_roll:", len(df.loc[df["Depth_interp_roll"].isnull()]))

#Plot
fig, axes = plt.subplots(nrows=2, figsize=(10, 6), sharex=True)

# Top subplot-Original depth, bad data, and yearly rolling mean
axes[0].plot(df["Year"], df["Depth"], label="Original Data", marker='o', linestyle='-', color='blue')
axes[0].plot(badyears, baddepths, label="Bad Data", marker='x', linestyle='none', color='red')
axes[0].plot(df["Year"], df["yearly_height"], label="Yearly Rolling Mean", linestyle='--', color='orange')
axes[0].set_title("Original Depth Data with Rolling Mean")
axes[0].set_ylabel("Depth (mm)")
axes[0].legend()

# Bottom subplot-Interpolated depth and interpolated rolling mean
axes[1].plot(df["Year"], df["Depth_interp"], label="Interpolated Data", marker='o', linestyle='-', color='green')
axes[1].plot(df["Year"], df["Depth_interp_roll"], label="Interpolated Rolling Mean", linestyle='--', color='purple')
axes[1].set_title("Interpolated Depth Data with Rolling Mean")
axes[1].set_xlabel("Year and Month")
axes[1].set_ylabel("Depth (mm)")
axes[1].legend()

# Layout and display
plt.tight_layout()
plt.show()

# D) The rolling mean interpolation describes the trends in missing depth data more accurately than the first because it accounts for the larger-scale trends than point-point connection.
````

````python
from scipy.stats import linregress

print("Part 1: ")
# Calculated the linear regression
slope, intercept, rvalue, pvalue, stderr = linregress(df["Year"], df["Depth_interp"])

# Calculated the linear regression fit
df["full_regression"] = slope * df["Year"] + intercept

print("Part 2: ")
# to subset recent data and calculate the linear regression
recent_df = df.loc[df["Year"] > 2007]
slope_new, intercept_new, rvalue_new, pvalue_new, stderr_new = linregress(
    recent_df["Year"], recent_df["Depth_interp"]
)

# to calculate the linear regression fit for recent data
df["recent_regression"] = slope_new * df["Year"] + intercept_new

# Plot
fig, axes = plt.subplots(nrows=2, figsize=(10, 6), sharex=True)

# Top subplot-Original depth, bad data, and yearly rolling mean
axes[0].plot(df["Year"], df["Depth"], label="Original Data", marker='o', linestyle='-', color='blue')
axes[0].plot(badyears, baddepths, label="Bad Data", marker='x', linestyle='none', color='red')
axes[0].plot(df["Year"], df["yearly_height"], label="Yearly Rolling Mean", linestyle='--', color='orange')
axes[0].set_title("Original Depth Data with Rolling Mean")
axes[0].set_ylabel("Depth (mm)")
axes[0].legend()

# Bottom subplot-Interpolated depth, rolling mean, full regression, and recent regression
axes[1].plot(df["Year"], df["Depth_interp"], label="Interpolated Data", marker='o', linestyle='-', color='green')
axes[1].plot(df["Year"], df["Depth_interp_roll"], label="Interpolated Rolling Mean", linestyle='--', color='purple')
axes[1].plot(df["Year"], df["full_regression"], label=f"Full Regression ({slope:.2f} mm/year)", linestyle='-.', color='orange')
axes[1].plot(df["Year"], df["recent_regression"], label=f"Recent Regression ({slope_new:.2f} mm/year)", linestyle=':', color='red')
axes[1].set_title("Interpolated Depth Data with Regressions")
axes[1].set_xlabel("Year and Month")
axes[1].set_ylabel("Depth (mm)")
axes[1].legend()

# Layout and display
plt.tight_layout()
plt.show()

# Print regression details
print(f"Full regression slope: {slope:.2f} mm/year")
print(f"Recent regression slope: {slope_new:.2f} mm/year")

# 1e) A linear model provides a good estimate of the overall trend in sea level rise as it was able to effectively provide a clear trend of change
# 2e) The slope of the recent regression is steeper than that of the full regression showing sea level is accelerating. Rate of increase in sea levels has been higher in recent years due to the accelerated ice melt and thermal expansion of seawater from global warming.
````
